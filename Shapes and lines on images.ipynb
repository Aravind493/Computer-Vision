```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
img1 = np.zeros(shape=[500,500,3], dtype=np.int32)
```

```python
plt.imshow(img1)
```

```python
cv2.rectangle(img1,pt1=(300,50),pt2=(400,100),color=(0,0,255),thickness=2)
plt.imshow(img1)
```

```python
cv2.rectangle(img1,pt1=(300,300),pt2=(400,400),color=(0,255,0),thickness=2)
plt.imshow(img1)
```

```python
cv2.circle(img1,center=(50,50),radius=25,color=(255,0,0),thickness=3)
plt.imshow(img1)
```

```python
cv2.circle(img1,center=(250,200),radius=25,color=(255,0,0),thickness=-1)
plt.imshow(img1)
```

```python
cv2.line(img1,pt1=(0,0),pt2=(500,500),color=(255,255,255),thickness=2)
plt.imshow(img1)
```
