```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
noise=np.random.randint(low=0,high=2,size=[500,500])
```

```python
noise.shape
```

```python
noise
```

```python
plt.imshow(noise)
```

```python
noise=noise*255
```

```python
noise
```

```python
img=np.zeros(shape=[500,500],dtype=np.int32)
plt.imshow(img,cmap='gray')
```

```python
cv2.circle(img,center=(250,250),radius=50,color=(255,0,0),thickness=5)
plt.imshow(img,cmap='gray')
```

```python
img=img+noise
```

```python
plt.imshow(img,cmap='gray')
```
